````python
from sklearn.metrics import precision_score, recall_score, f1_score, confusion_matrix
````

````python
# ตารางและคอลัมน์
target_col = 'Paror1Healthy0'
kfold_col = 'kfold'

# โมเดล
clf = RandomForestClassifier(n_estimators=100, random_state=42)
svm = SVC(kernel='linear')
Xgb = xgb.XGBClassifier()
````

````python
# รวมโมเดลไว้ใน dict
models = {
    "RandomForest": clf,
    "SVM (Linear)": svm,
    "XGBoost": Xgb
}
````

````python
results = {name: {
    "accuracy": [], "precision": [], "recall": [], "f1": [],
    "TP": [], "TN": [], "FP": [], "FN": []
} for name in models}

# ทำ 5-fold evaluation
for fold in range(1, 6):
    print(f"\n===== Fold {fold} =====")
    df_train = full[full[kfold_col] != fold]
    df_valid = full[full[kfold_col] == fold]

    # X และ y
    X_train = df_train.drop(columns=[target_col, kfold_col])
    y_train = df_train[target_col]
    X_valid = df_valid.drop(columns=[target_col, kfold_col])
    y_valid = df_valid[target_col]

    # วนลูปโมเดล
    for name, model in models.items():
        model.fit(X_train, y_train)
        preds = model.predict(X_valid)

        accuracy = accuracy_score(y_valid, preds)
        precision = precision_score(y_valid, preds, average='binary')
        recall = recall_score(y_valid, preds, average='binary')
        f1 = f1_score(y_valid, preds, average='binary')

        # confusion matrix: [[TN FP], [FN TP]]
        tn, fp, fn, tp = confusion_matrix(y_valid, preds).ravel()

        # บันทึก
        results[name]["accuracy"].append(accuracy)
        results[name]["precision"].append(precision)
        results[name]["recall"].append(recall)
        results[name]["f1"].append(f1)
        results[name]["TP"].append(tp)
        results[name]["TN"].append(tn)
        results[name]["FP"].append(fp)
        results[name]["FN"].append(fn)

        print(f"{name}: Accuracy = {accuracy:.4f}, Precision = {precision:.4f}, Recall = {recall:.4f}, F1 = {f1:.4f}")
        print(f"        TP={tp}, TN={tn}, FP={fp}, FN={fn}")
````

````python
print("\n===== Average Performance Across 5 Folds =====")
for name in models:
    avg_accuracy = sum(results[name]["accuracy"]) / 5
    avg_precision = sum(results[name]["precision"]) / 5
    avg_recall = sum(results[name]["recall"]) / 5
    avg_f1 = sum(results[name]["f1"]) / 5
    avg_tp = sum(results[name]["TP"]) / 5
    avg_tn = sum(results[name]["TN"]) / 5
    avg_fp = sum(results[name]["FP"]) / 5
    avg_fn = sum(results[name]["FN"]) / 5

    print(f"{name}:")
    print(f"  Accuracy = {avg_accuracy:.4f}, Precision = {avg_precision:.4f}, Recall = {avg_recall:.4f}, F1 Score = {avg_f1:.4f}")
    print(f"  Avg TP = {avg_tp:.1f}, TN = {avg_tn:.1f}, FP = {avg_fp:.1f}, FN = {avg_fn:.1f}")
````

#### 3.2 PCA 95% and 98% variance

````python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, f1_score, recall_score, precision_score, confusion_matrix
import numpy as np

variances_to_try = [0.95, 0.98]  # ค่า explained variance ที่ต้องการ
all_results = {}  # เก็บผลของทุก variance

for variance_target in variances_to_try:
    print(f"\n=== Explained Variance: {variance_target} ===")
    results = {name: [] for name in models.keys()}

    for fold in range(1, 6):
        train_df = full[full['kfold'] != fold].reset_index(drop=True)
        valid_df = full[full['kfold'] == fold].reset_index(drop=True)

        print(f"\nFold {fold}: Train shape = {train_df.shape}, Valid shape = {valid_df.shape}")

        X_train = train_df.drop(columns=['Paror1Healthy0', 'kfold']).values
        y_train = train_df['Paror1Healthy0'].values

        X_valid = valid_df.drop(columns=['Paror1Healthy0', 'kfold']).values
        y_valid = valid_df['Paror1Healthy0'].values

        scaler = StandardScaler()
        X_train_scaled = scaler.fit_transform(X_train)
        X_valid_scaled = scaler.transform(X_valid)

        # ทำ PCA โดยให้รักษา variance ตามที่กำหนด
        pca = PCA(n_components=variance_target)
        X_train_pca = pca.fit_transform(X_train_scaled)
        X_valid_pca = pca.transform(X_valid_scaled)

        # รายงานจำนวนคอลัมน์หลัง PCA
        n_components_used = X_train_pca.shape[1]
        print(f"  -> Components to reach {variance_target*100:.0f}% variance: {n_components_used} components")

        for name, model in models.items():
            model.fit(X_train_pca, y_train)
            y_pred = model.predict(X_valid_pca)

            acc = accuracy_score(y_valid, y_pred)
            f1 = f1_score(y_valid, y_pred)
            recall = recall_score(y_valid, y_pred)
            precision = precision_score(y_valid, y_pred)
            tn, fp, fn, tp = confusion_matrix(y_valid, y_pred).ravel()

            results[name].append({
                'fold': fold,
                'n_components': n_components_used,
                'accuracy': acc,
                'f1_score': f1,
                'recall': recall,
                'precision': precision,
                'tn': tn,
                'fp': fp,
                'fn': fn,
                'tp': tp
            })

    # สรุปผล
    for name in results:
        print(f"\n>>> Model: {name} | Explained Variance: {variance_target}")
        accuracies = []
        f1s = []
        recalls = []
        precisions = []
        tns, fps, fns, tps = [], [], [], []
        components_list = []

        for fold_result in results[name]:
            print(f" Fold {fold_result['fold']} - Components: {fold_result['n_components']}, "
                  f"Acc: {fold_result['accuracy']:.3f}, F1: {fold_result['f1_score']:.3f}, "
                  f"Recall: {fold_result['recall']:.3f}, Precision: {fold_result['precision']:.3f}")
            accuracies.append(fold_result['accuracy'])
            f1s.append(fold_result['f1_score'])
            recalls.append(fold_result['recall'])
            precisions.append(fold_result['precision'])
            tns.append(fold_result['tn'])
            fps.append(fold_result['fp'])
            fns.append(fold_result['fn'])
            tps.append(fold_result['tp'])
            components_list.append(fold_result['n_components'])

        print(f" Average  - Components: {np.mean(components_list):.1f}, "
              f"Acc: {np.mean(accuracies):.3f}, F1: {np.mean(f1s):.3f}, "
              f"Recall: {np.mean(recalls):.3f}, Precision: {np.mean(precisions):.3f}")
        print(f" Confusion Matrix Avg - TN: {np.sum(tns)/5:.1f}, FP: {np.sum(fps)/5:.1f}, "
              f"FN: {np.sum(fns)/5:.1f}, TP: {np.sum(tps)/5:.1f}")

    all_results[variance_target] = results
````
