### 3.1 Hoehn & Yahr severity estimation


#### 3.1.1 NO PCA

````python
status = df[['RegScale', 'kfold']]
status
````

````python
full = pd.concat([status,full], axis=1)
full
````

````python
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
import xgboost as xgb
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np
````

````python
target_col = 'RegScale'
kfold_col = 'kfold'

# โมเดล Regression
models = {
    "RandomForest": RandomForestRegressor(),
    "SVR": SVR(),
    "XGBoost": xgb.XGBRegressor()
}
````

````python
results = {name: {
    "r2": [], "mae": [], "mse": [], "rmse": [], "mape": [], "smape": []
} for name in models}

# ทำ 5-fold evaluation
for fold in range(1, 6):
    print(f"\n===== Fold {fold} =====")
    df_train = full[full[kfold_col] != fold]
    df_valid = full[full[kfold_col] == fold]

    X_train = df_train.drop(columns=[target_col, kfold_col])
    y_train = df_train[target_col]
    X_valid = df_valid.drop(columns=[target_col, kfold_col])
    y_valid = df_valid[target_col]

    for name, model in models.items():
        model.fit(X_train, y_train)
        preds = model.predict(X_valid)

        # คำนวณตัวชี้วัด
        r2 = r2_score(y_valid, preds)
        mae = mean_absolute_error(y_valid, preds)
        mse = mean_squared_error(y_valid, preds)
        rmse = np.sqrt(mse)
        mape = np.mean(np.abs((y_valid - preds) / (y_valid + 1e-8))) * 100  # กันหารด้วยศูนย์
        smape = np.mean(
            np.abs(preds - y_valid) / ((np.abs(y_valid) + np.abs(preds)) / 2 + 1e-8)
        ) * 100

        # เก็บผล
        results[name]["r2"].append(r2)
        results[name]["mae"].append(mae)
        results[name]["mse"].append(mse)
        results[name]["rmse"].append(rmse)
        results[name]["mape"].append(mape)
        results[name]["smape"].append(smape)

        print(f"{name}: R² = {r2:.4f}, MAE = {mae:.4f}, MSE = {mse:.4f}, "
              f"RMSE = {rmse:.4f}, MAPE = {mape:.2f}%, sMAPE = {smape:.2f}%")

# สรุปค่าเฉลี่ย
print("\n===== Average Performance Across 5 Folds =====")
for name in models:
    avg_r2 = np.mean(results[name]["r2"])
    avg_mae = np.mean(results[name]["mae"])
    avg_mse = np.mean(results[name]["mse"])
    avg_rmse = np.mean(results[name]["rmse"])
    avg_mape = np.mean(results[name]["mape"])
    avg_smape = np.mean(results[name]["smape"])

    print(f"{name}:")
    print(f"  R² = {avg_r2:.4f}, MAE = {avg_mae:.4f}, MSE = {avg_mse:.4f}, "
          f"RMSE = {avg_rmse:.4f}, MAPE = {avg_mape:.2f}%, sMAPE = {avg_smape:.2f}%")
````

#### 3.1.2 PCA 95% and 98% variance

````python
status = df[['RegScale', 'kfold']]
status
````

````python
full = pd.concat([status,full], axis=1)
full
````

````python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
````

````python
variances_to_try = [0.95, 0.98]
target_col = 'RegScale'
kfold_col = 'kfold'

# โมเดล Regression
models = {
    "RandomForest": RandomForestRegressor(),
    "SVR": SVR(),
    "XGBoost": xgb.XGBRegressor()
}
````

````python
all_results = {}

for variance_target in variances_to_try:
    print(f"\n=== Explained Variance: {variance_target} ===")
    results = {name: [] for name in models.keys()}

    for fold in range(1, 6):
        train_df = full[full[kfold_col] != fold].reset_index(drop=True)
        valid_df = full[full[kfold_col] == fold].reset_index(drop=True)

        print(f"\nFold {fold}: Train shape = {train_df.shape}, Valid shape = {valid_df.shape}")

        X_train = train_df.drop(columns=[target_col, kfold_col]).values
        y_train = train_df[target_col].values
        X_valid = valid_df.drop(columns=[target_col, kfold_col]).values
        y_valid = valid_df[target_col].values

        # สเกลข้อมูล
        scaler = StandardScaler()
        X_train_scaled = scaler.fit_transform(X_train)
        X_valid_scaled = scaler.transform(X_valid)

        # PCA
        pca = PCA(n_components=variance_target)
        X_train_pca = pca.fit_transform(X_train_scaled)
        X_valid_pca = pca.transform(X_valid_scaled)

        n_components_used = X_train_pca.shape[1]
        print(f"  -> Components to reach {variance_target*100:.0f}% variance: {n_components_used} components")

        for name, model in models.items():
            model.fit(X_train_pca, y_train)
            y_pred = model.predict(X_valid_pca)

            r2 = r2_score(y_valid, y_pred)
            mae = mean_absolute_error(y_valid, y_pred)
            mse = mean_squared_error(y_valid, y_pred)
            rmse = np.sqrt(mse)
            mape = np.mean(np.abs((y_valid - y_pred) / (y_valid + 1e-8))) * 100  # ป้องกันหารด้วยศูนย์
            smape = np.mean(
                np.abs(y_pred - y_valid) / ((np.abs(y_valid) + np.abs(y_pred)) / 2 + 1e-8)
            ) * 100

            results[name].append({
                'fold': fold,
                'n_components': n_components_used,
                'r2': r2,
                'mae': mae,
                'mse': mse,
                'rmse': rmse,
                'mape': mape,
                'smape': smape
            })

    # สรุปผลลัพธ์
    for name in results:
        print(f"\n>>> Model: {name} | Explained Variance: {variance_target}")
        r2s, maes, mses, rmses, mapes, smapes, components_list = [], [], [], [], [], [], []

        for fold_result in results[name]:
            print(f" Fold {fold_result['fold']} - Components: {fold_result['n_components']}, "
                  f"R²: {fold_result['r2']:.3f}, MAE: {fold_result['mae']:.3f}, "
                  f"MSE: {fold_result['mse']:.3f}, RMSE: {fold_result['rmse']:.3f}, "
                  f"MAPE: {fold_result['mape']:.2f}%, sMAPE: {fold_result['smape']:.2f}%")
            r2s.append(fold_result['r2'])
            maes.append(fold_result['mae'])
            mses.append(fold_result['mse'])
            rmses.append(fold_result['rmse'])
            mapes.append(fold_result['mape'])
            smapes.append(fold_result['smape'])
            components_list.append(fold_result['n_components'])

        print(f" Average  - Components: {np.mean(components_list):.1f}, "
              f"R²: {np.mean(r2s):.3f}, MAE: {np.mean(maes):.3f}, "
              f"MSE: {np.mean(mses):.3f}, RMSE: {np.mean(rmses):.3f}, "
              f"MAPE: {np.mean(mapes):.2f}%, sMAPE: {np.mean(smapes):.2f}%")

    # บันทึก
    all_results[variance_target] = results
````
