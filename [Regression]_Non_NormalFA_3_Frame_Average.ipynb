### 3.1 Hoehn & Yahr severity estimation

````python
full = pd.concat([status,full], axis=1)
full
````

````python

full_first = full[full['ForS'] == 1].reset_index(drop=True)
full_second = full[full['ForS'] == 2].reset_index(drop=True)
````

````python
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
import xgboost as xgb
import numpy as np
import pandas as pd
````

````python
target_col = 'RegScale'
kfold_col = 'kfold'

models = {
    "RandomForest": RandomForestRegressor(),
    "SVR": SVR(),
    "XGBoost": xgb.XGBRegressor()
}
````

````python
for label, df in [("full_first", full_first), ("full_second", full_second), ("full", full)]:
    print(f"\n### K-Fold Regression - {label} ###")

    results = {name: {
        "r2": [], "mae": [], "mse": [], "rmse": [], "mape": [], "smape": []
    } for name in models}

    for fold in range(1, 6):
        print(f"\n===== Fold {fold} =====")
        df_train = df[df[kfold_col] != fold].reset_index(drop=True)
        df_valid = df[df[kfold_col] == fold].reset_index(drop=True)

        drop_cols = [target_col, kfold_col]
        if 'ForS' in df.columns:
            drop_cols.append('ForS')

        X_train = df_train.drop(columns=drop_cols)
        y_train = df_train[target_col]
        X_valid = df_valid.drop(columns=drop_cols)
        y_valid = df_valid[target_col]

        for name, model in models.items():
            model.fit(X_train, y_train)
            preds = model.predict(X_valid)

            r2 = r2_score(y_valid, preds)
            mae = mean_absolute_error(y_valid, preds)
            mse = mean_squared_error(y_valid, preds)
            rmse = np.sqrt(mse)
            mape = np.mean(np.abs((y_valid - preds) / (y_valid + 1e-8))) * 100
            smape = np.mean(
                np.abs(preds - y_valid) / ((np.abs(y_valid) + np.abs(preds)) / 2 + 1e-8)
            ) * 100

            results[name]["r2"].append(r2)
            results[name]["mae"].append(mae)
            results[name]["mse"].append(mse)
            results[name]["rmse"].append(rmse)
            results[name]["mape"].append(mape)
            results[name]["smape"].append(smape)

            print(f"{name}: R² = {r2:.4f}, MAE = {mae:.4f}, "
                  f"MSE = {mse:.4f}, RMSE = {rmse:.4f}, MAPE = {mape:.2f}%, sMAPE = {smape:.2f}%")

    print("\n===== Average Performance Across 5 Folds =====")
    for name in models:
        avg_r2 = np.mean(results[name]["r2"])
        avg_mae = np.mean(results[name]["mae"])
        avg_mse = np.mean(results[name]["mse"])
        avg_rmse = np.mean(results[name]["rmse"])
        avg_mape = np.mean(results[name]["mape"])
        avg_smape = np.mean(results[name]["smape"])

        print(f"{name}:")
        print(f"  R² = {avg_r2:.4f}, MAE = {avg_mae:.4f}, "
              f"MSE = {avg_mse:.4f}, RMSE = {avg_rmse:.4f}, "
              f"MAPE = {avg_mape:.2f}%, sMAPE = {avg_smape:.2f}%")
````

#### 3.1.2 PCA 95% and 98% variance

````python
datasets = {
    "full_first": full_first,
    "full_second": full_second,
    "full": full
}

for label, df in datasets.items():
    print(f"\n### K-Fold + PCA - {label} ###")

    models = {
        "RandomForest": RandomForestRegressor(),
        "SVR": SVR(),
        "XGBoost": xgb.XGBRegressor()
    }

    for var in [0.95, 0.98]:
        print(f"\n===== PCA with Explained Variance = {var} =====")
        results = {name: {
            "r2": [], "mae": [], "mse": [], "rmse": [], "mape": [], "smape": [], "components": []
        } for name in models}

        for fold in range(1, 6):
            print(f"\n===== Fold {fold} =====")

            df_train = df[df[kfold_col] != fold].reset_index(drop=True)
            df_valid = df[df[kfold_col] == fold].reset_index(drop=True)

            drop_cols = [target_col, kfold_col]
            if 'ForS' in df.columns:
                drop_cols.append('ForS')

            X_train = df_train.drop(columns=drop_cols).values
            y_train = df_train[target_col].values
            X_valid = df_valid.drop(columns=drop_cols).values
            y_valid = df_valid[target_col].values

            # Scaling + PCA
            scaler = StandardScaler()
            X_train_scaled = scaler.fit_transform(X_train)
            X_valid_scaled = scaler.transform(X_valid)

            pca = PCA(n_components=var)
            X_train_pca = pca.fit_transform(X_train_scaled)
            X_valid_pca = pca.transform(X_valid_scaled)

            n_components_used = X_train_pca.shape[1]
            print(f"  -> Components to reach {var*100:.0f}% variance: {n_components_used} components")

            for name, model in models.items():
                model.fit(X_train_pca, y_train)
                preds = model.predict(X_valid_pca)

                r2 = r2_score(y_valid, preds)
                mae = mean_absolute_error(y_valid, preds)
                mse = mean_squared_error(y_valid, preds)
                rmse = np.sqrt(mse)
                mape = np.mean(np.abs((y_valid - preds) / (y_valid + 1e-8))) * 100  # ป้องกันหารด้วย 0
                smape = np.mean(
                    np.abs(preds - y_valid) / ((np.abs(y_valid) + np.abs(preds)) / 2 + 1e-8)
                ) * 100

                results[name]["r2"].append(r2)
                results[name]["mae"].append(mae)
                results[name]["mse"].append(mse)
                results[name]["rmse"].append(rmse)
                results[name]["mape"].append(mape)
                results[name]["smape"].append(smape)
                results[name]["components"].append(n_components_used)

                print(f"{name}: R² = {r2:.4f}, MAE = {mae:.4f}, "
                      f"MSE = {mse:.4f}, RMSE = {rmse:.4f}, MAPE = {mape:.2f}%, sMAPE = {smape:.2f}%")

        print(f"\n===== Average Performance Across 5 Folds (Explained Variance = {var}) =====")
        for name in models:
            avg_r2 = np.mean(results[name]["r2"])
            avg_mae = np.mean(results[name]["mae"])
            avg_mse = np.mean(results[name]["mse"])
            avg_rmse = np.mean(results[name]["rmse"])
            avg_mape = np.mean(results[name]["mape"])
            avg_smape = np.mean(results[name]["smape"])
            avg_components = np.mean(results[name]["components"])

            print(f"{name}:")
            print(f"  R² = {avg_r2:.4f}, MAE = {avg_mae:.4f}, MSE = {avg_mse:.4f}, "
                  f"RMSE = {avg_rmse:.4f}, MAPE = {avg_mape:.2f}%, sMAPE = {avg_smape:.2f}%")
            print(f"  Avg PCA Components Used: {avg_components:.1f}")
````
